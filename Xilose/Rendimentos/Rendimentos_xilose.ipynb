# Rendimento da produção de 3HP pelas vias de Xilose

```python
import matplotlib.pyplot as plt
import pandas as pd
from cobra.io import load_json_model
```

```python
# Modelos
modelobeta = load_json_model("saccharomycesXilose_Beta.json") # Beta-alanina
modelocoa = load_json_model("saccharomycesXilose_MaCoa.json") # Malonil-CoA
modelooxa = load_json_model("saccharomycesXilose_Oxal.json") # Oxaloacetato
modelomaoxa = load_json_model("saccharomycesXilose_MaOxal.json") # Malonil-CoA + Oxaloacetato
modeloPKbeta = load_json_model("saccharomycesXilose_PKBeta.json") # PK Beta
modeloPKcoa = load_json_model("saccharomycesXilose_PKMaCoa.json") # PK Malonil-CoA
modeloPKoxa = load_json_model("saccharomycesXilose_PKOxal.json") # PK Oxaloacetato
modeloPKmaoxa = load_json_model("saccharomycesXilose_PKMaOxal.json") # PK Malonil-CoA + Oxaloacetato
```

```python
# Dados de fluxo máximo
df1 = pd.read_csv("fluxo_maximo_beta.csv")
df2 = pd.read_csv("fluxo_maximo_macoa.csv")
df3 = pd.read_csv("fluxo_maximo_oxal.csv")
df4 = pd.read_csv("fluxo_maximo_maoxal.csv")
df5 = pd.read_csv("fluxo_maximo_pkbeta.csv")
df6 = pd.read_csv("fluxo_maximo_pkmacoa.csv")
df7 = pd.read_csv("fluxo_maximo_pkoxal.csv")
df8 = pd.read_csv("fluxo_maximo_pkmaoxal.csv")
```

```python
# Cálculo de rendimento e plot do gráfico
hp1 = float(df1.loc[df1['reaction'] == 'R_3HPxtO']['fluxes'])
xyl1 = float(df1.loc[df1['reaction'] == 'R_XYLxtI']['fluxes'])
hp2 = float(df2.loc[df2['reaction'] == 'R_3HPxtO']['fluxes'])
xyl2 = float(df2.loc[df2['reaction'] == 'R_XYLxtI']['fluxes'])
hp3 = float(df3.loc[df3['reaction'] == 'R_3HPxtO']['fluxes'])
xyl3 = float(df3.loc[df3['reaction'] == 'R_XYLxtI']['fluxes'])
hp4 = float(df4.loc[df4['reaction'] == 'R_3HPxtO']['fluxes'])
xyl4 = float(df4.loc[df4['reaction'] == 'R_XYLxtI']['fluxes'])
hp5 = float(df5.loc[df5['reaction'] == 'R_3HPxtO']['fluxes'])
xyl5 = float(df5.loc[df5['reaction'] == 'R_XYLxtI']['fluxes'])
hp6 = float(df6.loc[df6['reaction'] == 'R_3HPxtO']['fluxes'])
xyl6 = float(df6.loc[df6['reaction'] == 'R_XYLxtI']['fluxes'])
hp7 = float(df7.loc[df7['reaction'] == 'R_3HPxtO']['fluxes'])
xyl7 = float(df7.loc[df7['reaction'] == 'R_XYLxtI']['fluxes'])
hp8 = float(df8.loc[df8['reaction'] == 'R_3HPxtO']['fluxes'])
xyl8 = float(df8.loc[df8['reaction'] == 'R_XYLxtI']['fluxes'])

ox1 = float(df1.loc[df1['reaction'] == 'R_O2xtI']['fluxes'])
ox2 = float(df2.loc[df2['reaction'] == 'R_O2xtI']['fluxes'])
ox3 = float(df3.loc[df3['reaction'] == 'R_O2xtI']['fluxes'])
ox4 = float(df4.loc[df4['reaction'] == 'R_O2xtI']['fluxes'])
ox5 = float(df5.loc[df5['reaction'] == 'R_O2xtI']['fluxes'])
ox6 = float(df6.loc[df6['reaction'] == 'R_O2xtI']['fluxes'])
ox7 = float(df7.loc[df7['reaction'] == 'R_O2xtI']['fluxes'])
ox8 = float(df8.loc[df8['reaction'] == 'R_O2xtI']['fluxes'])

fig, ax = plt.subplots(1, 1, figsize=(13, 5), dpi=100)
ax2 = ax.twinx()

a = ax.bar(["Beta-Alanina","Malonil-CoA","Oxaloacetato","Via Dupla","PK-Beta","PK-Malonil-CoA","PK-Oxaloacetato","PK-Via Dupla"], [-hp1/xyl1, -hp2/xyl2, -hp3/xyl3, -hp4/xyl4, -hp5/xyl5, -hp6/xyl6, -hp7/xyl7, -hp8/xyl8])
ax.set_xlabel("Vias de produção")
ax.set_ylabel("Rendimento (3HP mmol/g DW h)")
a[0].set_color('lightgreen')
a[1].set_color('yellow')
a[2].set_color('blue')
a[3].set_color('pink')
a[4].set_color('green')
a[5].set_color('orange')
a[6].set_color('darkblue')
a[7].set_color('deeppink')

ax.bar_label(a)
ax2.scatter(["Beta-Alanina","Malonil-CoA","Oxaloacetato","Via Dupla","PK-Beta","PK-Malonil-CoA","PK-Oxaloacetato","PK-Via Dupla"], [-ox1,-ox2,-ox3,-ox4,-ox5,-ox6,-ox7, -ox8], color='#ff0000')
ax2.set_ylabel("Consumo (O2 mmol/g DW h)")
plt.title("Rendimento das vias de produção de 3HP por Xilose", fontsize = 15)
plt.savefig("Rendimento_xilose.png")
plt.show()
```
