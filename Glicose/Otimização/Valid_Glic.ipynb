```python
import matplotlib.pyplot as plt
import pandas as pd

# Data
labels = [
    'Tipo selvagem',
    'ACC1 super',
    'ACC1 + TPS3 super',
    'ACC1 + TPS3 + TRP5 super',
    'ACC1 super + SDH3 (KO)',
    'ACC1 + TRP5 super + SDH3 (KO)',
    'ACC1 + TPS3 + TRP5 super + SDH3 (KO)'
]

biomassa_values = [0.97194, 0.8190, 0.81230, 0.80559, 0.72534, 0.71844, 0.71003] #Fluxo biomassa
_3hp_values = [0, 430.21253/10, 431.29346/10, 432.55456/10, 446.96704/10, 448.22813/10, 449.66938/10] #Rendimento (3HP/Fluxo Glicose)

fig, ax1 = plt.subplots(figsize=(16, 8))

# Eixo Glicose
ax1.set_xlabel('Conjunto de genes', fontsize=14)
ax1.set_ylabel('Produção de Biomassa (h-1)', color='tab:red', fontsize=14)
ax1.set_xticks(range(len(labels)))
ax1.set_xticklabels(labels, rotation=45, ha="right", fontsize=12)
ax1.plot(labels, biomassa_values, color='tab:red', marker='o', label="Produção de biomassa")
ax1.tick_params(axis='y', labelcolor='tab:red')

# Eixo 3HP
ax2 = ax1.twinx()
ax2.set_ylabel('Rendimento de 3HP (mg/g glicose)', color='tab:blue', fontsize=14) 
ax2.plot(labels, _3hp_values, color='tab:blue', marker='s', label="Produção de 3HP")
ax2.tick_params(axis='y', labelcolor='tab:blue')

ax2.axhline(y = 3320.27508/10, label = 'Rendimento máximo teórico 3HP', color = 'black', linestyle = 'dashed') 

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')
plt.title('Relação crescimento e produção de 3HP para conjuntos de genes', fontsize=16)
plt.tight_layout()
plt.show()
```
