```python
import pandas as pd
import matplotlib.pyplot as plt
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
```

```python
df = pd.read_csv('ROU_Xyl.txt', sep=";")
df
```

```python
output_notebook()

xs, ys = [], []

for index, row in df.iterrows():
    a = row['BCPYxWYIELD']
    x, y = a.strip('[').strip(']').split(',')
    x = float(x)
    y = float(y)
    xs.append(x)
    ys.append(y)

p = figure(title='SPEA2 Sub-Super expressão e KO Xilose', x_axis_label='BCPY', y_axis_label='WYIELD', plot_width=600, plot_height=600)
p.title.text_font_size = '16pt'
p.xaxis.axis_label_text_font_size = '14pt'
p.yaxis.axis_label_text_font_size = '14pt'
p.scatter(xs, ys, size=8, color="navy", alpha=0.5)
show(p)
```
